```python
headers = {"Client-Identifier": "ZainebDMProject"}

url = "https://gbfs.urbansharing.com/bikemi.com/station_status.json"
response = requests.get(url, headers=headers, timeout=10)
data = response.json()

stations = data["data"]["stations"]

df_status = pd.DataFrame(stations)
df_status.head()
```

```python
df_status["collection_timestamp"] = pd.to_datetime(data["last_updated"], unit="s")
```

```python
url_info = "https://gbfs.urbansharing.com/bikemi.com/station_information.json"
info = requests.get(url_info, headers=headers).json()

df_info = pd.DataFrame(info["data"]["stations"])
df_info.head()
```

```python
df_merged = df_status.merge(df_info, on="station_id", how="left")
df_merged.head()
```

```python
cols_to_save = [
    "station_id", "name", "lat", "lon", "num_bikes_available",
    "num_docks_available", "num_vehicles_available",
    "collection_timestamp"
]

df_merged[cols_to_save].to_csv("bikemi_status.csv", index=False)
```
